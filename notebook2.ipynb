# Revisión de dataset con columnas únicas

```python
import pandas as pd
```

```python
data = pd.read_csv('latimUniqueCols.csv')
```

```python
data.head()
```

```python
data.info()
```

## Identificación de valores únicos en var_category

```python
data.var_category.unique()
```

## Generación de nuevos datasets a partir de cada categoría

```python
df_dict = {name: data.loc[data['var_category'] == name] for name in data.var_category.unique()}
```

```python
df_dict['Employment']
```

```python
df_dict['Population']
```

```python
df_dict['Unemployment']
```

```python
df_dict['Poverty and vulnerability']
```

```python
df_dict
```

## Guardando los nuevos dataframes a archivos

```python
for key in df_dict:
    name = 'latim' + key.replace(' ','') + '.csv'
    df_dict[key].to_csv(name, index=False)
```
