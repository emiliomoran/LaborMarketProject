```python
import pandas as pd
```

```python
df = pd.read_csv('analystFinal2.csv')
```

```python
def parseSalario(s):
    s2 = s.replace('$','')
    commaIdx = s2.find(',')
    final = s2[:commaIdx]
    #print(final)
    #return int(final)
    return final
```

```python
df['Salario'] = df['Salario'].map(parseSalario)
```

```python
df['Salario'] = pd.to_numeric(df['Salario'])
```

```python
df['Jornada'] = df['Jornada'].apply(lambda x: x.upper())
```

```python
def parseDate(d):
    idx = d.find("-")
    dParse = d[:idx]
    return dParse
```

```python
df['Activo desde'] = df['Activo desde'].map(parseDate)
```

```python
df['Activo desde'] = pd.to_datetime(df['Activo desde'], dayfirst=True)
```
